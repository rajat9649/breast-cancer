```python
for i in num:
    plt.figure()
    sns.boxplot(df[i])
```

```python
for i in num:
    plt.figure()
    sns.boxplot(y=df[i],x=df['diagnosis'])
```

```python
for i in num:
    plt.figure()
    sns.set_style('whitegrid')
    sns.distplot(df[i],bins=50)
```

```python
m = df.loc[df['diagnosis'] == 'M']
b = df.loc[df['diagnosis'] == 'B']

for i in num:
    plt.figure()
    sns.distplot(m[[i]],hist=True,label='maligant')
    sns.distplot(b[[i]],hist=True,label='brest')
    plt.legend(loc='best')
```
